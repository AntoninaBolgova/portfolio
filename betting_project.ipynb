```python
#data types changing,  replace ',' to '.' to change data type to float
dat3['GameDate']=pd.to_datetime(dat3['GameDate'])
dat3['SumRevA']=(dat3['SumRevA']).str.replace(',','.')
dat3['SumRevB']=(dat3['SumRevB']).str.replace(',','.')
dat3['SumRevA']=(dat3['SumRevA']).astype(float)
dat3['SumRevB']=(dat3['SumRevB']).astype(float)
```

```python
#counting client revenue
dat3['ClientRev']=dat3['SumRevA']+dat3['SumRevB']
```

```python
#creating new column
df2['cashback']=0
```

```python
#counting cashback. cashback % depends on client revenue.Each paying user takes part in the bonus program: from 0 to
#$100 of revenue per month – 5% cashback, from 101 to 1000 – 10%, from 1001 to 10000 – 15%, more than 10001 – 20%
for row in range(len(df2)):
    if df2.iloc[row,9]>10001:
        df2.iloc[row,10]=df2.iloc[row,3]*0.20
    elif df2.iloc[row,9]>1001:
        df2.iloc[row,10]=df2.iloc[row,3]*0.15
    elif df2.iloc[row,9]>101:
        df2.iloc[row,10]=df2.iloc[row,3]*0.10
    elif df2.iloc[row,9]>0:
        df2.iloc[row,10]=df2.iloc[row,3]*0.05
    
    
```

```python
#creating table with clients who paid deposit 
clientsact['year'] = clientsact['year'].astype(dtype = 'Int64')
clientsact['month'] = clientsact['month'].astype(dtype = 'Int64')
total['year'] = total['year'].astype(dtype = 'Int64')
total['month'] = total['month'].astype(dtype = 'Int64')
conv=clientsact.merge(total,left_on=['year','month'], right_on=['year','month'], how = 'outer')
```

```python
#Conversion Rate: Deposit Registration = % of all registered players who made their first real money deposit.
conv['conv'] = round((conv['clientspaid']/conv['clientspermonth'])*100).astype(dtype  ='Int64')
conv
```
